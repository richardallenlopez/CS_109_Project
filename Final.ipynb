###### Loading and Cleaning Data

Function to remove the left handed star in the name string and add lefty indicator column

```python
def remove_stars(rowdata):
    if rowdata['pitcher_name'][-1] == '*':
        rowdata['pitcher_name'] = rowdata['pitcher_name'][:-1]
    return rowdata
```

Load at-bat, pitch, and WAR data

```python
# // define constants and lists to iterate over
years = ['2011','2012','2013']
files = ['war2011','war2012','war2013']

merge_list = ['url', 'num','inning','next_','inning_side','gameday_link']
joined = []
war_data = []

# // define innings limit to threshold at (and separate out starters)
innings_threshold = 80

for x, f in enumerate(files):
    
    # // Load at-bat data
    atbat_temp = pd.read_csv('../Scraped Data/'+years[x]+'/atbat.csv');
    # // Load pitch data
    pitch = pd.read_csv('../Scraped Data/'+years[x]+'/pitch.csv');

    # // Remove the 'atbat.' that starts every column
    atbat_temp.rename(columns=lambda x: x[6:], inplace=True)
    # // Remove the 'pitch.' that starts every column
    pitch.rename(columns=lambda x: x[6:], inplace=True)

    # // Merge the dataframes on the overlapping data
    join = pd.merge(pitch, atbat_temp, how='inner', on=merge_list)

    # // Add year label onto the end of Pitcher's Names
    join['pitcher_name'] = join['pitcher_name']+years[x]
    
    # // Add data frame into list to store it
    joined.append(join)

    # // === // === // === // === // === // === 
    
    # // Load WAR data, thresholding pitchers appopriately
    war = pd.read_csv('../Scraped Data/'+years[x]+'/'+ f +'.csv');
    war = war[war.IP > innings_threshold]
    
    # // Subset the Data Frame
    war_sub = war[['Name','WAR']]

    # // Rename the columns for merging purposes
    war_sub.rename(columns={'Name': 'pitcher_name'}, inplace=True)

    # // Eliminate traded pitchers
    bool_war = (war_sub.groupby('pitcher_name').size() == 1)
    war_sub = war_sub.groupby('pitcher_name').mean()[bool_war].reset_index().apply(remove_stars,axis=1)
    
    # // Append year to names
    war_sub['pitcher_name'] = war_sub['pitcher_name']+years[x]
    
    war_data.append(war_sub)
    
joined = joined[0].append(joined[1:])
    
war_data = war_data[0].append(war_data[1:])
```

#### Searching for Features that Correlate with WAR

Now that we know the general distribution of WAR data among pitchers, we want to look for correlations between different properties of pitchers/pitches and WAR. Specifically, we are going to ignore all outcome related data and just look at pitch trajectory information. It is obvious that good outcomes (e.g. fewer hits, less runs, more called strikes, etc.) would result in a pitcher with a higher WAR. We want to use the things that pitchers have more control over and see if any of those characteristics give a higher WAR.

It is important to note here that a univariate correlation does not really tell us much about how a variable truly affects WAR, but rather the multivariate correlation. For this reason, we seek to find a good mix of features that demonstrate univariate correlation with WAR as well as intuitive correlate with WAR. 

###### Total Pitches

A lot of the things we are going to try and test need information on total pitches, so we are going to get that information now.

```python
# how many pitches did the pitcher throw?
total_pitch = war_b4_14[['pitcher_name','WAR']].groupby('pitcher_name').count()
total_pitch = total_pitch.reset_index().rename(columns={'WAR':'total_pitch'})

plt.scatter(total_pitch.total_pitch,Y.WAR)
plt.xlabel("Number of Pitches")
plt.ylabel("WAR")
plt.title("WAR v. Number of Pitches")
plt.show()
print "Pearson R is : " + str(scipy.stats.pearsonr(total_pitch.total_pitch,Y.WAR)[0])
```

Here, we are using the Pearson R value to quantify how well a particular dataset correlates. One can read more about the Pearson Correlation Coefficient on wikipedia (http://en.wikipedia.org/wiki/Pearson_product-moment_correlation_coefficient), bt in general it shoule be noted at an R = 1 is complete linear correlation, R = 0 is complete absence of correlation, and R = -1 is complete negative correlation. 

So this is a prety great R value! But we need to think about why this is true and make sure it makes sense ot use it to predict. The reason why better pitchers are pitching more is because the coachers are leaving them in longer because they are pitching well. The number of pitches does not actually predict how well a pitcher is, but how well a pitcher is would predict how many pitches they throw in the season. We need to make sure we are thinking through out results in order to make sure that we are using the right information to predict how good a pitcher is.

###### Proportion of Pitches in the Strike Zone

We think that if a pitcher pitches more pitches in the strike zone, they might be a better pitcher. Pitch F/X already stores location information under the $\texttt{zone}$ column, and so we can take advantage of that to find out if a given pitch is in the strike zone.

```python
# // Get the Number of Strikes per pitcher
strikes = war_b4_14[(war_b4_14.zone.isin(range(1,10)))].groupby('pitcher_name').count()
strikes = strikes.reset_index().rename(columns={'type':'strikes'})[['pitcher_name','strikes']]

# // Get the Percentage of pitches that are strikes (by dividing by total pitch)
strike_percent = pd.merge(total_pitch,strikes,on='pitcher_name',how='left')
strike_percent['strike_percent'] = strike_percent['strikes']/strike_percent['total_pitch']
strike_percent = strike_percent[['pitcher_name','strike_percent']]

# // Plot the result
plt.scatter(strike_percent['strike_percent'],Y['WAR'])
plt.xlabel('Percent of Pitches in the Strike Zone')
plt.ylabel('WAR')
plt.title("WAR v. Percent of Pitches in the Strike Zone")
plt.show()
print "Pearson R is : " + str(scipy.stats.pearsonr(strike_percent['strike_percent'],Y['WAR'])[0])
```

This correlates decently well. Let's keep on looking.

###### Pitches not in the Center of the Strike Zone

Another feature of interest is to count the number of pitches not directly in the center of the strike zone that otherwise were in the strike zone. Again, we can capitalize on the $\texttt{zone}$ column from the Pitch F/X data in order to count this quantity.

```python
not_middle = war_b4_14[war_b4_14['zone'].isin([1,2,3,4,6,7,8,9])].groupby('pitcher_name').count()
not_middle = not_middle.reset_index().rename(columns={'zone':'not_middle'})

not_middle = pd.merge(strikes,not_middle,on='pitcher_name',how='left')
not_middle['not_middle'] = not_middle['not_middle']/not_middle['strikes']

plt.scatter(not_middle['not_middle'],Y['WAR'])
plt.xlabel('Proportion of Corners')
plt.ylabel('WAR')
plt.title("WAR v. Percent of Pitches not in the Middle")
print "Pearson R is : " + str(scipy.stats.pearsonr(not_middle['not_middle'],Y['WAR'])[0])
```

It's definitely interesting that this feature negatively correlates with WAR. This seems to conflict with the idea that a pitcher should not throw down the middle of the plate (although obviously doesn't prove anything).

###### Variance of Mean of Start Speeds for each Pitch (with thresholding constraints)

This idea, similar to the prior one, is a metric to quantify how much a pitcher mixes up his speeds across his entire arsenal. The idea is that if a batter has to mentally account for pitches coming in at variable speeds, it will be harder to have a good at-bat.

```python
# // threshold the available values
pitch_threshold = 50
bool_list = [war_b4_14.groupby(['pitcher_name','pitch_type']).count().reset_index().des > pitch_threshold]

# // take the mean of values for pitches thrown 
result = war_b4_14.groupby(['pitcher_name','pitch_type']).mean().reset_index()[np.transpose(bool_list)]

# // take the variance for each pitcher and record
result = result.groupby("pitcher_name").var().reset_index()

plt.scatter(result.start_speed,Y.WAR)
plt.xlabel('Variance of Start Speed')
plt.ylabel('WAR')
plt.title("WAR vs. Variance of Mean Pitch Speed (of a Pitcher's Arsenal)")

print "Pearson R is : " + str(scipy.stats.pearsonr(result.start_speed,Y.WAR)[0])
```

#### Clustering

Since none of the above values have any significantly strong correlation with WAR, we now try to cluster pitcher based on pitcher "style". The idea here is that different pitchers rely on different strategies to get batters out. For example Tim Wakefield or RA Dickey utitlize slow, unpredicatable knucklesballs to get batters out, and so their average fastball speed probably doesn't matter at all. On the other hand, pitchers like Aroldis Chapman thrive on their 100+ mph fastballs in order to get batters out. The hope is that, in clustering pitchers by pitcher style, we can find stronger correlations.

First, we seek to eliminate all non-starting pitcher and cluster only on the subset of pitchers in which we are interested.

```python
# // Create list of validu pitchers
starterList = Y.pitcher_name.tolist()

# // subset data frame accordingly
war_sub = war_b4_14[war_b4_14.pitcher_name.isin(starterList)]

# // Count the total number of pitches for each pitcher
total_count = pd.groupby(war_sub,'pitcher_name').count().reset_index()
total_count = total_count[['pitcher_name','des']].rename(columns={'des':'total'})
```

As per the above discussion, it's probably useful to cluster based on pitch selection. Here, we calculate the percentage of pitcher thrown that were fastballs or breaking pitches.

```python
# // calculate percentage fastball thrown
grouped_count = pd.groupby(war_sub,['pitcher_name','pitch_type']).count().reset_index()[['pitcher_name','pitch_type','des']]
grouped_count = grouped_count

fb_count = pd.groupby(grouped_count[grouped_count.pitch_type.isin(['FF','FT','FC','FS','SI'])],'pitcher_name').sum().reset_index()
fb_count.rename(columns={'des':'fb'}, inplace=True)

count_joined = pd.merge(fb_count,total_count,on=['pitcher_name'],how='inner')
count_joined['proportion'] = count_joined.fb / count_joined.total;

fb_percent = count_joined.drop(['fb','total'],1)
```

```python
# // calculate percentage breaking pitch thrown
grouped_count = pd.groupby(war_sub,['pitcher_name','pitch_type']).count().reset_index()
grouped_count = grouped_count[['pitcher_name','pitch_type','des']]

breaking_count = pd.groupby(grouped_count[grouped_count.pitch_type.isin(['SL','CB','CU','KC'])],'pitcher_name').sum().reset_index()
breaking_count.rename(columns={'des':'breaking'}, inplace=True)

count_joined = pd.merge(breaking_count,total_count,on=['pitcher_name'],how='inner')
count_joined['proportion'] = count_joined.breaking / count_joined.total;

breaking_percent = count_joined.drop(['breaking','total'],1).fillna(0)
```

Another common thing for announcers to comment on is how some pitcher's "pitch to contact" while others go for more swing-and-misses. We can cluster pitchers based on these attributes by counting the general outcome of their pitches.

```python
# // calculate percentage of pitches put in play
grouped_count = pd.groupby(war_sub,['pitcher_name','type']).count().reset_index()
grouped_count = grouped_count[['pitcher_name','type','des']]

in_play_count = grouped_count[grouped_count.type.isin(['X'])]
in_play_count.rename(columns={'des':'in_play'}, inplace=True)

count_joined = pd.merge(in_play_count,total_count,on=['pitcher_name'],how='inner')
count_joined['proportion'] = count_joined.in_play / count_joined.total;

in_play_percent = count_joined.drop(['in_play','total','type'],1)
```

We can do the same for the percentage of pitches that were called a strike for whatever reason (e.g. called strikes, foul, swinging strikes, etc).

```python
# // calculate percentage of strikes not put in play
grouped_count = pd.groupby(war_sub,['pitcher_name','type']).count().reset_index()
grouped_count = grouped_count[['pitcher_name','type','des']]

strike_count = grouped_count[grouped_count.type.isin(['S'])]
strike_count.rename(columns={'des':'strike'}, inplace=True)

count_joined = pd.merge(strike_count,total_count,on=['pitcher_name'],how='inner')
count_joined['proportion'] = count_joined.strike / count_joined.total;

strike_percent_clust = count_joined.drop(['strike','total','type'],1)
```

In the same vein, we may be interested in clustering based on whether pitchers tend to get their outs as flyball or groundball pitchers. 

```python
# // calculate percentage of groundball vs groundballs/flyballs
g_vs_f = pd.groupby(war_sub[war_sub.event.isin(['Groundout','Flyout'])],['pitcher_name','event']).count().reset_index().iloc[:,0:3]
tot_num = pd.groupby(g_vs_f,'pitcher_name').sum().reset_index().rename(columns={'atbat_des':'c'})

# tot_num_sub = tot_num[tot_num.pitcher_name.isin(pitcher_list)]
g_vs_f_sub = g_vs_f[g_vs_f.pitcher_name.isin(starterList)]

g_num = g_vs_f_sub[g_vs_f_sub.event == 'Groundout'].rename(columns={'atbat_des':'c'})

# ground_proportion = np.divide(g_num.pitcher.as_matrix().astype(float),tot_num_sub.pitcher.as_matrix())
ground_proportion = np.divide(g_num.c.as_matrix().astype(float),tot_num.c.as_matrix())

ground_percent = strike_percent_clust.copy()
ground_percent = ground_percent.drop('proportion',1)
ground_percent['proportion'] = ground_proportion
```

The last thing that we will cluster on is the "entropy" of a given pitcher's arsenal. This idea can be read about at http://en.wikipedia.org/wiki/Entropy_(information_theory). The overall idea is that the "Entropy" of a Pitcher's Arsenal is greater if they more evenly spread out their pitch choice. Thus pitcher's that tend to mix up their pitch selection more often will have a greater value of Entropy.

```python
# // get 'bits' of pitcher (related to predictability of arsenal)
total_pitch = war_b4_14[['pitcher_name','WAR']].groupby('pitcher_name').count().reset_index().rename(columns={'WAR':'total_pitch'})

def divstuff(row_data):
    pitcher_name = row_data.pitcher_name
    N = total_pitch[total_pitch.pitcher_name == pitcher_name].total_pitch
    row_data['pi'] = row_data.tfs/float(N)
    row_data['logp'] = np.log(row_data.tfs/float(N))
    row_data['bits'] = -1*row_data['pi']*row_data['logp']/float(np.log(N))
    return row_data

bits = war_b4_14.groupby(['pitcher_name','pitch_type']).count().reset_index().apply(divstuff, axis=1)
bits = bits.groupby('pitcher_name').sum().reset_index()[['pitcher_name','bits']]
```

```python
bits.head()
```

We can take advantage of these calculated numbers to cluster pitchers. First, we write a function to normalize the data in order to make sure that we are treating variation among each pitcher with equal weight.

```python
def normalize(X):
    subtracted = np.tile(np.mean(X,0),(X.shape[0],1))
    X_centered = X - subtracted
    divided = np.tile(np.std(X_centered,axis=0),(X.shape[0],1))
    X_norm = np.divide(X_centered,divided)
    return X_norm, divided, subtracted
```

We can now normalize the data and format it nicely into a numpy matrix, for analysis purposes.

```python
# // Calculate the normalized data matrix for these features
explore = pd.concat([fb_percent,
                     breaking_percent.proportion,
                     in_play_percent.proportion,
                     strike_percent_clust.proportion,
                     ground_percent.proportion,
                     bits.bits],axis=1)
# // need to fill nan because some pitcher don't throw a breaking pitch (Rodney, Valverde, Rivera, Dickey)
X = explore.iloc[:,1::].fillna(0).as_matrix()

# // normalize the data set
X_norm, divided, subtracted = normalize(X)
```

We cluster the data into 3 groups (a parameter which can be varied) and examine the separation of each category. The cluster-object  $\texttt{cluster_fitted}$ can be used to categorize pitchers from later years (i.e. 2014) at a later point.

```python
clusterer = sklearn.cluster.KMeans(n_clusters=3, max_iter = 1000, n_init=10)
clusterer_fitted = clusterer.fit(X_norm)
labels = clusterer_fitted.predict(X_norm)

# // Get list of names from one of the existing DataFrames.
cluster_frame = ground_percent.copy();

cluster_frame['cluster'] = labels
cluster_frame = cluster_frame.drop('proportion',1)
```

Now that we have this information, we can create our own cross-validation function to see how well we can predict if a pitcher is above or below average based on the cluster and our features. 

###### Cross Validation

First, lets try to do cross validation with the clusters that we have and see how well that works

```python
# CLUSTERING

# Cross Validation Scores
cross_scores = []
cross_scores2 = []

# Create our own cross validation function
for z in range(1,25):
    
    # Initialize correctness
    totalcorrect = []
    total = []
    totalcorrect2 = []
    total2 = [] 

    # Cross validation
    for i in range(len(x)):
        kf = sklearn.cross_validation.KFold(len(x[i]), n_folds=10)
        correct = []
        count = []
        correct2 = []
        count2 = []
        
        for train_i, test_i in kf:
            x_val, x_testing = x[i][train_i], x[i][test_i]
            y_val, y_testing = y[i][train_i], y[i][test_i]

            # train Classifier
            forest = sklearn.ensemble.RandomForestClassifier(n_estimators=z)
            forest.fit(x_val,y_val)
            knn = sklearn.neighbors.KNeighborsClassifier(n_neighbors=z,weights='uniform')
            knn.fit(x_val,y_val)

            # predict the class
            y_hat = forest.predict(x_testing)
            y_hat2 = knn.predict(x_testing)


            # See how many we got right
            correct.append(sum(y_testing == y_hat))
            count.append(len(y_testing))
            correct2.append(sum(y_testing == y_hat2))
            count2.append(len(y_testing))

        # Append that to the total correct
        totalcorrect.append(correct)
        total.append(count)
        totalcorrect2.append(correct2)
        total2.append(count2)

    # Make np array
    totalcorrect = np.array(totalcorrect)
    total = np.array(total)
    totalcorrect2 = np.array(totalcorrect2)
    total2 = np.array(total2)

    # Compare correctness to total
    totalcorrect = np.sum(totalcorrect,axis=0)
    total = np.sum(total,axis=0)
    totalcorrect2 = np.sum(totalcorrect2,axis=0)
    total2 = np.sum(total2,axis=0)

    cross_scores.append(totalcorrect/(total*1.0))
    cross_scores2.append(totalcorrect2/(total2*1.0))
    
# Plot
sns.boxplot(cross_scores)
plt.title('CV Scores with Different Forest Sizes')
plt.xlabel('Number of Trees in Forest')
plt.ylabel('Cross Validation Scores')
plt.axhline(y=0.5, c='green')
plt.show()

sns.boxplot(cross_scores2)
plt.title('CV Scores with Different # Neighbors')
plt.xlabel('Number of Neighbors')
plt.ylabel('Cross Validation Scores')
plt.axhline(y=0.5, c='green')
plt.show()
```

## Final Analysis

We initially embarked on this data exploration seeking to quantify a pitcher's "stuff" and determine how well that "stuff" can predict their effectiveness as a player. Specifically, using the Pitch F/X data set, we narrowed our attention to the movement, speed, and trajectory information of a pitcher's arsenal, independent of the outcomes.
After significant quantities of in-depth analysis, we sort of have to conclude that, conditioning on the fact that the pitchers are a major league starters, "stuff" does not really exist. This is evident from the ineffective prediction capabilities of our classifier, which attempts to determine whether a pitcher will be in the upper or lower half of pitcher WAR ratings. Indeed, it is far more useful to consider a pitcher's past performance in order to illuminate future performance -- knowledge of their "stuff" can only serve to amplify that knowledge, not as a basis.

###### (Spurious?) Correlations

Along the way, however, we found a lot of interesting correlations and disproved a fair amount of myths for baseball. For example, ground ball pitchers (who have a groundout-to-flyout ratio > 1) do not necessarily tend to have greater WAR than flyball pitchers. Also interestingly, pitchers that hit the corners of the plate more frequency only improve marginally in performance compared to other pitchers that pound the strike zone. A baseball team on a budget could easily capitalize on gaps of other organization's knowledge to find undervalued players based on these discoveries.

###### Clustering

Another interesting idea in this project was to try to group pitcher's according to their pitching "style". Announcers in baseball frequently comment on how some pitchers "pitch-to-contact" while others have "swing-and-miss stuff". We were interested in exploring these clusters in order to improve our predictions of how good pitchers are, but the analysis is of baseball interest on its own.

We found that, as expected, clustering pitchers based on pitch style does in fact improve our prediction capacity, agreeing with the notion that different pitchers rely on different style to get batters out. 

In general, it seems that the best separator of pitcher styles is pitch selection, with (as expected) pitchers that throw their fastball more having greater correlation between fastball speed and WAR. 

###### Machine Learning

We tried to find the best features possible that would indicate a pitcher with high WAR. However, even with the best features that we could find, our cross validation scores still only ended up predicting at 60% success rate, which is only 10 points increase from the baseline. This is not particularly helpful, particularly in light of the amount of money that these teams spend on these players. The fact that clustering did improve our preditions does verify that pitchers rely on different styles to find their success. 

A suggestion for a future direction for this section is to include WAR data from 2013 and see if we can improve the prediction capability of pitcher success over just that data. The idea here is that the most naive approach to picking players is to examine their prior year's WAR and then evaluate based on that. If we can improve even slightly over this baseline value, then perhaps this method may have a place in terms of influencing baseball decisions. 

It seems from this analysis that, given that a pitcher is a major league starter, the actually quality of their "stuff" does not particularly determine their overall success for the year. We thus suggest more strongly weighting the actual performance information of a pitcher in determining their value. However, if the project were extended as suggested, then perhaps a combination of stats and stuff could serve to better predict pitcher performance better than either of them along.
